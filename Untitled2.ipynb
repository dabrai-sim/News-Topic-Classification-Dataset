###Define the preprocessing steps: tokenization, lemmatization, stopword removal, and cleansing.

```python
# Initialize SpaCy for lemmatization
nlp = spacy.load('en_core_web_sm')

# Function for data cleansing
def preprocess_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenization and lemmatization
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if token.text.lower() not in stopwords.words('english')]
    return ' '.join(tokens)

# Apply preprocessing to the training and testing data
train_df['cleaned_text'] = train_df['Description'].apply(preprocess_text)
test_df['cleaned_text'] = test_df['Description'].apply(preprocess_text)
```

###Use CountVectorizer and TfidfVectorizer to vectorize the cleaned text.

```python
# Vectorization
X_train_count = CountVectorizer().fit_transform(train_df['cleaned_text'])
X_train_tfidf = TfidfVectorizer().fit_transform(train_df['cleaned_text'])

X_test_count = CountVectorizer().fit(train_df['cleaned_text']).transform(test_df['cleaned_text'])
X_test_tfidf = TfidfVectorizer().fit(train_df['cleaned_text']).transform(test_df['cleaned_text'])

y_train = train_df['Class Index']
y_test = test_df['Class Index']
```

###Train and evaluate the models using CountVectorizer and TfidfVectorizer.

```python
# Function to evaluate models
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title(f'Confusion Matrix for {model.__class__.__name__}')
    plt.show()
```

```python
# Logistic Regression
print("Logistic Regression with CountVectorizer:")
evaluate_model(LogisticRegression(C=0.1, solver='liblinear'), X_train_count, X_test_count, y_train, y_test)
```

```python
print("Logistic Regression with TFIDFVectorizer:")
evaluate_model(LogisticRegression(C= 1, solver ='liblinear'), X_train_tfidf, X_test_tfidf, y_train, y_test)
```
